```python
import pandas as pd
import numpy as np

year = "2018"

df = pd.read_excel(f"../../data/pollution/raw/{year}/excel/{year}_PM10_1g.xlsx")
```

```python
df = df.set_index("Nr")
```

```python
df = df.transpose()
```

```python
import requests

station_data = requests.get("https://api.gios.gov.pl/pjp-api/v1/rest/station/findAll?size=500")
```

```python
station_infos = station_data.json()["Lista stacji pomiarowych"]
```

```python
station_infos
```

```python
sensor_infos = pd.read_csv("../../allSensors.csv")
```

```python
sensor_infos
```

```python
unknown_stations = []
for _, row in df.iterrows():
    station_name = row["Kod stacji"]
    data = row.iloc[5:]
    station_info = [station for station in station_infos if station["Kod stacji"] == station_name]
    if len(station_info) == 0:
        unknown_stations.append(station_name)
    else:
        station_id = station_info[0]["Identyfikator stacji"]
        data_df = pd.DataFrame({"date": data.index, "value": data.values})
        data_df = data_df.set_index("date")
        data_df["value"].replace("", np.nan, inplace=True)
        data_df.dropna(subset=["value"], inplace=True)
        data_df.to_csv(f"../../data/pollution/raw/{year}/pm10/{year}_{station_id}.csv")

unknown_stations
```
