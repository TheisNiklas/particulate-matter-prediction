### Visualise the maximum and minimum values of the stations per year

Intended to analyse the existence of outlier in the data.

The max values with the highest one 1.019 all seem to be valid. \
The min values are at 0 or less for some stations so those are not valid. \
Therefore the decision is to interpolate all values <= 0.


```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df2019 = pd.read_json("../../data/pollution/raw/2019/pm10/meta_data.json")
df2020 = pd.read_json("../../data/pollution/raw/2020/pm10/meta_data.json")
df2021 = pd.read_json("../../data/pollution/raw/2021/pm10/meta_data.json")

fig = make_subplots(column_titles=["Max values"], x_title="station_index", y_title="pm10")

fig.add_trace(go.Bar(x=df2019.index, y=df2019["max"], name="Max 2019"))
fig.add_trace(go.Bar(x=df2020.index, y=df2020["max"], name="Max 2020"))
fig.add_trace(go.Bar(x=df2021.index, y=df2021["max"], name="Max 2021"))

fig.update_layout(height=600)
```

```python
fig = make_subplots(column_titles=["Min values"], x_title="station_index", y_title="pm10")
fig.add_trace(go.Bar(x=df2019.index, y=df2019["min"], name="Min 2019"))
fig.add_trace(go.Bar(x=df2020.index, y=df2020["min"], name="Min 2020"))
fig.add_trace(go.Bar(x=df2021.index, y=df2021["min"], name="Min 2021"))
```
