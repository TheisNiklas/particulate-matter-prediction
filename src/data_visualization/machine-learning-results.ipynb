### Creates plots of the training and validation metrics of the initial runs


```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
```

```python
def load_data(filename):
    df = pd.read_csv(
        filename,
        names=["epoch", "timestamp", "value"],
        index_col="epoch",
    )
    df.index = df.index.astype(int).__add__(1)
    return df
```

```python
train_mae_IMT_11 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-11__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_12 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-12__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_13 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-13__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_14 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-14__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_15 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-15__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_16 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-16__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_17 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-17__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_19 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-19__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_20 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-20__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_21 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-21__training_train_epoch_mean_absolute_error.csv")
train_mae_IMT_22 = load_data("../../data/model-training/results/initial-model-tests/train-mae/IMT-22__training_train_epoch_mean_absolute_error.csv")
```

```python
val_mae_IMT_11 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-11__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_12 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-12__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_13 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-13__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_14 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-14__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_15 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-15__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_16 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-16__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_17 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-17__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_19 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-19__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_20 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-20__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_21 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-21__training_validation_epoch_mean_absolute_error.csv")
val_mae_IMT_22 = load_data("../../data/model-training/results/initial-model-tests/val-mae/IMT-22__training_validation_epoch_mean_absolute_error.csv")
```

```python
fig = make_subplots(column_titles=["Training Mean Absolute Error"])
fig.update_layout(height=750)

fig.add_trace(go.Scatter(x=train_mae_IMT_11.index, y=train_mae_IMT_11["value"], name="base_run", line={"width":3}))
fig.add_trace(go.Scatter(x=train_mae_IMT_12.index, y=train_mae_IMT_12["value"], name="lstm_size_64"))
fig.add_trace(go.Scatter(x=train_mae_IMT_13.index, y=train_mae_IMT_13["value"], name="two_lstm_size_32"))
fig.add_trace(go.Scatter(x=train_mae_IMT_14.index, y=train_mae_IMT_14["value"], name="two_lstm_size_64"))
fig.add_trace(go.Scatter(x=train_mae_IMT_15.index, y=train_mae_IMT_15["value"], name="bidirectional_lstm"))
fig.add_trace(go.Scatter(x=train_mae_IMT_16.index, y=train_mae_IMT_16["value"], name="bidirectional_two_lstm"))
fig.add_trace(go.Scatter(x=train_mae_IMT_17.index, y=train_mae_IMT_17["value"], name="average_pooling"))
fig.add_trace(go.Scatter(x=train_mae_IMT_19.index, y=train_mae_IMT_19["value"], name="two_dense_layer"))
fig.add_trace(go.Scatter(x=train_mae_IMT_20.index, y=train_mae_IMT_20["value"], name="dropout_0.4"))
fig.add_trace(go.Scatter(x=train_mae_IMT_21.index, y=train_mae_IMT_21["value"], name="dropout_0.5"))
fig.add_trace(go.Scatter(x=train_mae_IMT_22.index, y=train_mae_IMT_22["value"], name="dropout_0.6"))
```
