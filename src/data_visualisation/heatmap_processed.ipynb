#### Prepair the dataframes used for plotting and plot a seperate Heatmap per year.


```python
# 2019
result19 = [
    result_jan19,
    result_feb19,
    result_mae19,
    result_apr19,
    result_mai19,
    result_jun19,
    result_jul19,
    result_aug19,
    result_sep19,
    result_okt19,
    result_nov19,
    result_dez19,
]

result19_df = pd.concat(result19)
result19_df = result19_df.to_numpy().ravel()
dates19 = pd.date_range("1/1/2019", periods=365, freq="D")
events19 = pd.Series(result19_df, index=dates19)
events19.dropna(inplace=True)

# 2020
result20 = [
    result_jan20,
    result_feb20,
    result_mae20,
    result_apr20,
    result_mai20,
    result_jun20,
    result_jul20,
    result_aug20,
    result_sep20,
    result_okt20,
    result_nov20,
    result_dez20,
]

result20_df = pd.concat(result20)
result20_df = result20_df.to_numpy().ravel()
dates20 = pd.date_range("1/1/2020", periods=366, freq="D")
events20 = pd.Series(result20_df, index=dates20)
events20.dropna(inplace=True)

# 2021
result21 = [
    result_jan21,
    result_feb21,
    result_mae21,
    result_apr21,
    result_mai21,
    result_jun21,
    result_jul21,
    result_aug21,
    result_sep21,
    result_okt21,
    result_nov21,
    result_dez21,
]

result21_df = pd.concat(result21)
result21_df = result21_df.to_numpy().ravel()
dates21 = pd.date_range("1/1/2021", periods=365, freq="D")
events21 = pd.Series(result21_df, index=dates21)
events21.dropna(inplace=True)

# 2022
result22 = [
    result_jan22,
    result_feb22,
    result_mae22,
    result_apr22,
    result_mai22,
    result_jun22,
    result_jul22,
    result_aug22,
    result_sep22,
    result_okt22,
    result_nov22,
    result_dez22,
]

result22_df = pd.concat(result22)
result22_df = result22_df.to_numpy().ravel()
dates22 = pd.date_range("1/1/2022", periods=365, freq="D")
events22 = pd.Series(result22_df, index=dates22)
events22.dropna(inplace=True)

cmap = "RdYlGn_r"

calplot.calplot(events19, cmap=cmap, dropzero=True)

calplot.calplot(events20, cmap=cmap, dropzero=True)

calplot.calplot(events21, cmap=cmap, dropzero=True)

calplot.calplot(events22, cmap=cmap, dropzero=True)
```
